```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
d = pd.read_csv('Melbourne_housing_FULL.csv')
d.head()
```

```python
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Propertycount', 
               'Distance', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = d[cols_to_use]
df.head(2)
```

```python
df.isna().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)

df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df.Suburb = le.fit_transform(df.Suburb)
df.Type = le.fit_transform(df.Type)
df.SellerG = le.fit_transform(df.SellerG)
df.Method = le.fit_transform(df.Method)
```

```python
x = df
y = df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## Using L1 Regularization

```python
from sklearn import linear_model
```

```python
l1 = linear_model.Lasso(alpha = 50, max_iter = 100, tol = 0.1)
l1.fit(X_train, y_train)
```

```python
l1.score(X_test, y_test)
```

## Using L2 Regularization

```python
from sklearn.linear_model import Ridge
```

```python
l2 = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
l2.fit(X_train,y_train)
```

```python
l2.score(X_test,y_test)
```
