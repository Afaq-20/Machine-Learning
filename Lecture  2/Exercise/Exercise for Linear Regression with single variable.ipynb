```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('canada income.csv')
```

```python
df.head()
```

```python
plt.xlabel('Year')
plt.ylabel('per capital income (US$)')
plt.plot(df.year,df['per capita income (US$)'])
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
def predictValue(value):
    return reg.predict(np.array([[value]]))
```

```python
predictValue(2020)
```
