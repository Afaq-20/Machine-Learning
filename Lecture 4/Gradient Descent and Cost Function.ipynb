```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradDescent(x,y):
    m = b = 0
    rate = 0.01
    n = len(x)
    for i in range(999):   
        Ypredicted = m * x + b

        plt.plot(x,Ypredicted,color = 'red')
        
        md = -(2/n) * sum(x * (y-Ypredicted))
        bd = -(2/n) * sum((y-Ypredicted))

        m = m - rate * md
        b = b - rate * bd
```

```python
gradDescent(x,y)
```
