```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('homeprices.csv')
```

```python
df
```

```python
dummy = pd.get_dummies(df.town)
dummy = dummy.astype('int')
```

```python
merg = pd.concat([df,dummy], axis = 'columns')
merg
```

```python
final = merg.drop(['town','west windsor'], axis = 'columns')
```

```python
final
```

```python
x = final.drop(['price'], axis = 'columns')
y = final.price
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[2800,0,0]])
```

```python
print(f'Accuracy percentile of our model: {(model.score(x,y))*100}') 
```

## Now i will do same operation with Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dfle = df
dfle.town = le.fit_transform(dfle.town)
```

```python
dfle
```

```python
X = dfle[['town', 'area']]
```

```python
Y = dfle.price
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Assuming 'categorical_features' is the index of your categorical column
categorical_features = [0]

preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'  # Pass through numerical columns without transformation
)

# Fit and transform your data using the preprocessor
X = preprocessor.fit_transform(X)
```

```python
X = X[:,1:]
```

```python
X
```

```python
model.fit(X,Y)
```

```python
model.predict([[1,0,2800]])
```
